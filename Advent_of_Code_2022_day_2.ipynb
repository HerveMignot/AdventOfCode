## Part 1

```
# A / rock -> Y / paper    +2
# B / paper -> Z / scissor +3
# C / scissors -> X / rock +1
map_hand = {'A': 'X', 'B': 'Y', 'C': 'Z', 'X': 'A', 'Y': 'B', 'Z': 'C'}
best_hand = {'A': 'Y', 'B': 'Z', 'C': 'X'}
shape_score = {'A': 1, 'B': 2, 'C': 3, 'Y': 2, 'Z': 3, 'X': 1}
```

### Short exploration

```
lost = []
draw = []
win = []

for g in games:
  if best_hand[g[0]] == g[1]:
    win.append(g)
  elif map_hand[g[0]] == g[1]:
    draw.append(g)
  else:
    lost.append(g)

print(len(win), len(draw), len(lost))
print(len(win) * 6 + len(draw) *3)
```

```
score = 0

for g in games:
  if best_hand[g[0]] == g[1]:
    score += 6 + shape_score[g[1]]
  elif map_hand[g[0]] == g[1]:
    score += 3 + shape_score[g[1]]
  else:
    score += 0 + shape_score[g[1]]

score
```

```
# X means you need to lose, Y means you need to end the round in a draw, and Z means you need to win
# X lost, Y draw, Z win
# A / rock -> Y / paper    +2
# B / paper -> Z / scissor +3
# C / scissors -> X / rock +1

# Adding score makes things easier, but not really nice.
# But a two level dict would make it... 'X': {'play': {'A': 'Z',...}, 'score': 0}

strategy_guide = {
    'X': {'A': 'Z', 'B': 'X', 'C': 'Y', 'score': 0},
    'Y': {'A': 'X', 'B': 'Y', 'C': 'Z', 'score': 3},
    'Z': {'A': 'Y', 'B': 'Z', 'C': 'X', 'score': 6},
    }
```

```
sum([strategy_guide[g[1]]['score'] + shape_score[strategy_guide[g[1]][g[0]]] for g in test_games])
```
