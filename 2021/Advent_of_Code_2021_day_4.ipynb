## Code

```python
import io

import numpy as np

def read_inputs(input: str):
  with io.StringIO(input) as f:
    lines = f.readlines()

  draws = list(map(int, lines[0][:-1].split(',')))

  boards = list()
  for start in range(2, len(lines), 6):
    boards.append(np.array([list(map(int, l[:-1].split())) for l in lines[start:start + 5]]))
  
  return draws, boards
```

```python
draws, boards = read_inputs(INPUT)
```

```python
test_draws, test_boards = read_inputs(TESTS)
```

## Part 1

```python
def check_draw(board, check, value):
  """Mark on the check array the value if it is on the board"""
  loc = np.argwhere(board == value) 
  if loc.shape[0] == 0:
    return check
  else:
    check[loc[0][0], loc[0][1]] = 1
    return check

def has_row_or_col_complete(check, size=5):
  """Test if the check array has one column or row completed"""
  return (check.sum(axis=0).max() == size) or (check.sum(axis=1).max() == size)
```

### Implement on tests

```python
checks = [np.zeros(board.shape) for board in test_boards]

score = None
found = False

for ball in test_draws:
  for board, check in zip(test_boards, checks):
    if has_row_or_col_complete(check_draw(board, check, ball)):
      found = True
      break
  
  if found:
    score = (board.sum() - np.multiply(board, check).sum()) * ball
    break

if score:
  print('Score:', score)
  print(ball, test_draws)
  print(board)
  print(check)
else:
  print('No grid finished')
```

### Puzzle input

```python
checks = [np.zeros(board.shape) for board in boards]

score = None
found = False

for ball in draws:
  for board, check in zip(boards, checks):
    if has_row_or_col_complete(check_draw(board, check, ball)):
      found = True
      break
  
  if found:
    score = (board.sum() - np.multiply(board, check).sum()) * ball
    break

if score:
  print('Score:', score)
  print(board)
  print(check)
else:
  print('No grid finished')
```

## Part 2

### Implement on tests

```python
checks = [np.zeros(board.shape) for board in test_boards]

score = None
complete = set()

for ball in test_draws:
  for number, (board, check) in enumerate(zip(test_boards, checks)):
    if has_row_or_col_complete(check_draw(board, check, ball)):
      complete.add(number)
      if len(complete) == len(test_boards):
        score = (board.sum() - np.multiply(board, check).sum()) * ball
        break
  if score:
    break

if score:
  print('Score:', score)
  print(ball, test_draws)
  print(board)
  print(check)
else:
  print('No grid finished')
```

### Puzzle input

```python
checks = [np.zeros(board.shape) for board in boards]

score = None
complete = set()

for ball in draws:
  for number, (board, check) in enumerate(zip(boards, checks)):
    if has_row_or_col_complete(check_draw(board, check, ball)):
      complete.add(number)
      if len(complete) == len(boards):
        score = (board.sum() - np.multiply(board, check).sum()) * ball
        break
  if score:
    break

if score:
  print('Score:', score)
  print(ball)
  print(board)
  print(check)
else:
  print('No grid finished')
```
