```python
def load_hands(puzzle: str) -> list:
  return [(hand, int(bid)) for hand, bid in map(str.split, puzzle.splitlines())]
```

```python
load_hands(TEST)
```

## Part 1

```python
ranks = ['11111', '2111', '221', '311', '32', '41', '5']
CARDS = ['2', '3', '4', '5', '6', '7', '8', '9', 'T', 'J', 'Q', 'K', 'A'] #['A', 'K', 'Q', 'J', 'T', '9', '8', '7', '6', '5', '4', '3', '2']

def get_hand_rank(hand: str, cards=CARDS) -> str:
  return str(ranks.index(''.join(map(str, sorted([hand.count(card) for card in set(hand)], reverse=True))))) + ''.join(map(lambda c: chr(65 + cards.index(c)), hand))
```

## Part 1 Test

```python
# 32T3K is the only one pair and the other hands are all a stronger type, so it gets rank 1.
# KK677 and KTJJT are both two pair. Their first cards both have the same label, but the second card of KK677 is stronger (K vs T), so KTJJT gets rank 2 and KK677 gets rank 3.
# T55J5 and QQQJA are both three of a kind. QQQJA has a stronger first card, so it gets rank 5 and T55J5 gets rank 4.

sum([bid * rank for bid, rank in map(lambda e: (e[1][2], e[0]), enumerate(sorted([(hand, get_hand_rank(hand), bid) for hand, bid in load_hands(TEST)], key=lambda elem: elem[1]), start=1))])
```

```python
CARDS_P2 = ['J', '2', '3', '4', '5', '6', '7', '8', '9', 'T', 'Q', 'K', 'A']
```

```python
def use_joker(hand: str) -> str:
  return hand.replace('J', sorted([(card, hand.count(card) * 100 + CARDS_P2.index(card)) for card in (set(hand) - {'J'})], key=lambda e: e[1], reverse=True)[0][0]) if hand != 'JJJJJ' else hand
```

```python
def get_hand_rank_P2(hand: str, cards) -> str:
  new_hand = use_joker(hand)
  return str(ranks.index(''.join(map(str, sorted([new_hand.count(card) for card in set(new_hand)], reverse=True))))) + ''.join(map(lambda c: chr(65 + cards.index(c)), hand))
```
