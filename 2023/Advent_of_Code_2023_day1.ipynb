```
SPELLED = {'one': '1', 'two': '2', 'three': '3', 'four': '4', 'five': '5',
           'six': '6', 'seven': '7', 'eight': '8', 'nine': '9'} #, 'zero': '0'}

def convert(s: str) -> str:
  res, i = '', 0
  while i < len(s):
    found = False
    for digit in SPELLED:
      if s[i:].startswith(digit):
        res += s[i:].replace(digit, SPELLED[digit])[0]
        found = True
        i += len(digit)
        break
    if not found:
      res += s[i]
      i += 1
  return res

assert convert('two1nine') == '219'
assert convert('eightwothree') == '8wo3'
assert convert('abcone2threexyz') == 'abc123xyz'
assert convert('xtwone3four') == 'x2ne34'
assert convert('4nineeightseven2') == '49872'
assert convert('zoneight234') == 'z1ight234'
assert convert('7pqrstsixteen') == '7pqrst6teen'
```

```
def search_digits(s: str) -> int:

  def is_digit(s: str) -> str:
    if '0' <= s[0] <= '9':
      return s[0]
    for digit in SPELLED:
      if s.startswith(digit):
        return SPELLED[digit]

  left, right = None, None
  for i in range(len(s)):
    if res := is_digit(s[i:]):
      left = res
      break
  for i in range(-1, -len(s) - 1, -1):
    if res := is_digit(s[i:]):
      right = res
      break

  return int(left + right)
```

```
assert search_digits('two1nine') == 29
assert search_digits('eightwothree') == 83
assert search_digits('abcone2threexyz') == 13
assert search_digits('xtwone3four') == 24
assert search_digits('4nineeightseven2') == 42
assert search_digits('zoneight234') == 14
assert search_digits('7pqrstsixteen') == 76
```
