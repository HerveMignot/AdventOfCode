# Advent Of Code 2024 - Day 1

## Solution part 1

```
def puzzle_1(input: str) -> int:
  l1, l2 = map(list, zip(*[(int(couple[0]), int(couple[1])) for couple in [line.split("   ") for line in input.splitlines()]]))
  return sum([abs(couple[0] - couple[1]) for couple in zip(sorted(l1), sorted(l2))])
```

## Solution part 2

```
import collections

def puzzle_2(input: str) -> int:
  l1, l2 = map(list, zip(*[(int(couple[0]), int(couple[1])) for couple in [line.split("   ") for line in input.splitlines()]]))
  counts = collections.Counter(l2)
  return sum([e * counts[e] for e in l1])
```

```
test = """3   4
4   3
2   5
1   3
3   9
3   3
"""
```
