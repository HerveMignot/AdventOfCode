# Advent of code - day 2

```python
TEST_PUZZLE = "11-22,95-115,998-1012,1188511880-1188511890,222220-222224,1698522-1698528,446443-446449,38593856-38593862,565653-565659,824824821-824824827,2121212118-2121212124"
```

## Solution part 1

```python
def puzzle1(input: str) -> int:
  ranges, total = input.split(","), 0
  for rng in ranges:
    start, end = rng.split("-")
    for id in range(int(start), int(end) + 1):
      sid = str(id)
      if len(sid) % 2 == 0:
        if sid[:len(sid) // 2] == sid[len(sid) // 2:]:
          total += id
  return total
```

```python
assert puzzle1(TEST_PUZZLE) == 1227775554, "Not ok"
```

```python
puzzle1(INPUT_PUZZLE)
```

## Solution part 2

```python
def puzzle2(input: str) -> int:
  ranges, total = input.split(","), 0
  for rng in ranges:
    start, end = rng.split("-")
    for id in range(int(start), int(end) + 1):
      sid = str(id)
      for size in range(2, len(sid) + 1):
        if sid[:len(sid) // size] * size == sid:
          total += id
          break
  return total
```

```python
assert puzzle2(TEST_PUZZLE) == 4174379265, "Not yet"
```

```python
puzzle2(INPUT_PUZZLE)
```
